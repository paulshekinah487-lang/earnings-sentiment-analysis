## 03 - Earnings Call Transcripts

This notebook manages our earnings call transcript collection.

### What is a transcript?
Every quarter, public companies host an earnings call where the CEO and CFO
discuss financial results. These calls are recorded and transcribed publicly.

### What we collect?
- Transcripts from 2018 to 2025
- Focus on CEO and CFO speaking sections only
- 5 key market periods: pre-COVID, COVID crash, recovery, AI boom, tariff shock

### Why only CEO and CFO sections?
Analyst Q&A introduces outside voices. We want to measure only what
company leadership chose to say — that's where sentiment signals live.

```python
# Import libraries
import os
import pandas as pd

# Define our 5 key earnings calls to start with
# One from each major market period 2018-2025

key_calls = [
    {"ticker": "AAPL", "quarter": "Q1_2020", "date": "2020-01-28", "period": "Pre-COVID"},
    {"ticker": "AAPL", "quarter": "Q2_2020", "date": "2020-04-30", "period": "COVID Crash"},
    {"ticker": "MSFT", "quarter": "Q4_2022", "date": "2022-07-26", "period": "Rate Hike Peak"},
    {"ticker": "NVDA", "quarter": "Q2_2023", "date": "2023-08-23", "period": "AI Boom"},
    {"ticker": "AMZN", "quarter": "Q1_2025", "date": "2025-05-01", "period": "Tariff Shock"},
]

df_calls = pd.DataFrame(key_calls)
print(df_calls)
```
