## 02 - Company List & Earnings Dates

This notebook builds our master list of 50 S&P 500 companies across 5 sectors.

### Why these companies?
We select 10 companies per sector to ensure our analysis is not biased
towards one industry. Different sectors behave differently during market
cycles — technology booms while energy struggles, or vice versa.

### Sectors covered
- Technology
- Finance
- Healthcare
- Consumer
- Energy

```python
# Import libraries
import pandas as pd
import yfinance as yf

# Our 50 companies across 5 sectors
companies = {
    "ticker": [
        # Technology (10)
        "AAPL", "MSFT", "GOOGL", "META", "NVDA",
        "INTC", "CRM", "ADBE", "AMD", "ORCL",
        # Finance (10)
        "JPM", "BAC", "WFC", "GS", "MS",
        "BLK", "AXP", "C", "USB", "PNC",
        # Healthcare (10)
        "JNJ", "PFE", "UNH", "MRK", "ABT",
        "TMO", "DHR", "AMGN", "MDT", "BMY",
        # Consumer (10)
        "AMZN", "WMT", "HD", "MCD", "NKE",
        "SBUX", "TGT", "COST", "PG", "KO",
        # Energy (10)
        "XOM", "CVX", "COP", "SLB", "EOG",
        "PXD", "MPC", "VLO", "OXY", "HAL"
    ],
    "sector": (
        ["Technology"] * 10 +
        ["Finance"] * 10 +
        ["Healthcare"] * 10 +
        ["Consumer"] * 10 +
        ["Energy"] * 10
    )
}

df = pd.DataFrame(companies)
print(df)
print(f"\nTotal companies: {len(df)}")
```

```python
# Save company list to CSV
df.to_csv("../data/company_list.csv", index=False)

print("Company list saved successfully!")
```

```python
# Pull stock data for all 50 companies
# We go from 2018 to 2025 - our full research window

failed = []

for ticker in df["ticker"]:
    try:
        stock = yf.Ticker(ticker)
        data = stock.history(start="2018-01-01", end="2025-12-31")
        data.to_csv(f"../data/prices/{ticker}_2018_2025.csv")
        print(f"✓ {ticker} saved")
    except:
        failed.append(ticker)
        print(f"✗ {ticker} failed")

print(f"\nDone! Failed tickers: {failed}")
```

```python
# Replace PXD with another energy company
stock = yf.Ticker("DVN")  # Devon Energy
data = stock.history(start="2018-01-01", end="2025-12-31")
data.to_csv("../data/prices/DVN_2018_2025.csv")
print("✓ DVN saved")
```

```python
# Update company list - replace PXD with DVN
df["ticker"] = df["ticker"].replace("PXD", "DVN")
df.to_csv("../data/company_list.csv", index=False)
print("Company list updated!")
```
