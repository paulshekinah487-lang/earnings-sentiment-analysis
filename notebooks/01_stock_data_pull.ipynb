```python
# Import the libraries we need
import yfinance as yf
import pandas as pd

print("Libraries loaded successfully!")
```

```python
# Pull Apple's stock price data around Q1 2020 earnings call
# Apple reported earnings on January 28, 2020
# We pull 90 days after to see what happened to the stock

ticker = "AAPL"
stock = yf.Ticker(ticker)

data = stock.history(start="2020-01-28",end="2020-04-28")

print(data.head(10))
```

```python
# Save the data to a csv file in our prices folder
data.to_csv("../data/prices/AAPL_Q1_2020.csv")

print("File saved successfully")
```

This notebook pulls historical stock price data for S&P 500 companies
using the yfinance library.

## What it does
- Connects to Yahoo Finance using a company ticker symbol
- Pulls daily price data for a specified date range
- Saves the data as a CSV file for later analysis
