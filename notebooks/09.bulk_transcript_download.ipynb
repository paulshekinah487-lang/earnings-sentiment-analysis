## 09 - Bulk Transcript Download from Hugging Face

This notebook downloads earnings call transcripts for all 50 companies
from the Hugging Face S&P 500 Earnings Transcripts dataset.

### Why Hugging Face?
Instead of manually copying 89 transcripts, we download them
programmatically. This is faster, more accurate, and more impressive
to show on GitHub.

### What we download?
- 2 transcripts per company * 50 companies = 100 transcripts
- Quarter 1: Q1 2020 (Pre-COVID baseline)
- Quarter 2: Q1 2025 if available, else Q2 2023, else Q4 2022

### What stays the same?
All 14 existing transcripts are untouched.
We only add new ones alongside them.

```python
# Import libraries
from datasets import load_dataset
import pandas as pd
import os

print("Libraries imported successfully!")
print("Loading dataset from Hugging Face...")
print("This may take a few minutes - it's a large dataset...")
```

```python
# Load the dataset from Hugging Face
# We stream it so we don't download the entire thing at once
ds = load_dataset(
    "kurry/sp500_earnings_transcripts",
    split="train",
    streaming=True
)

print("Dataset connected successfully!")
print("Streaming mode - we only download what we need")
```
